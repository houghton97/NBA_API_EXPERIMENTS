```python
from nba_api.stats.endpoints import DraftCombineStats
from nba_api.stats.library.parameters import SeasonAll
import pandas as pd

# initialize list to hold dataframes
dfs = []
last_years = ['2018', '2019', '2020', '2021', '2022']
for year in last_years:
    dcs = DraftCombineStats(season_all_time=year)
    df = dcs.get_data_frames()[0]
# add year column
    df['Year'] = year
    dfs.append(df)

combined_df = pd.concat(dfs, ignore_index=True)

# print combined dataframe
combined_df.to_csv('draft_combine_original.csv', sep=';', decimal=',')

```

```python
df = combined_df
```

```python
cols_remove= [col for col in df.columns if col.endswith('_FT_IN') or col.startswith('OFF_DRIB') or col.startswith('ON_MOVE') or col.startswith('SPOT_')]
cols_remove += ['Year', 'BENCH_PRESS', 'FIRST_NAME', 'LAST_NAME']
```

```python
df = df.drop(columns=cols_remove)
```

```python
df
```

```python
height_columns = ['HEIGHT_WO_SHOES', 'HEIGHT_W_SHOES', 'WINGSPAN', 'STANDING_REACH', 'HAND_LENGTH', 'HAND_WIDTH',\
     'STANDING_VERTICAL_LEAP', 'MAX_VERTICAL_LEAP']
weight_columns = ['WEIGHT']
```

```python
for column in height_columns:
    df[column] = df[column].astype(float) * 0.0254
    df[column] = df[column].round(2)
# Convierte las columnas de peso a kg
for column in weight_columns:
    df[column] = df[column].astype(float) * 0.453592
    df[column] = df[column].round(2)
```

```python
import numpy as np
df = df.replace(np.nan, '-', regex=True)
```
